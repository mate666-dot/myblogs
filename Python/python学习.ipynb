**用matplotlib绘制等高线(contour)图形**  

*数据初始化*

````python
import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

#%matplotlib inline

#定义等高线高度函数
def f(x,y):
    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

#数据数目
n=256
#定义自变量取值范围
x=np.linspace(-3,3,n)
y=np.linspace(-3,3,n)

#生成网格数据
X,Y=np.meshgrid(x,y)

#设置图像大小尺寸
plt.figure(figsize=(10,6))

#填充等高线颜色，8是等高线分为几部分
plt.contour(X,Y,f(X,Y),8,alpha=0.75,cmap=plt.cm.hot)
````

````python
#填充等高线颜色 
plt.figure(figsize=(10,6))
plt.contourf(X,Y,f(X,Y),8,alpha=0.75,cmap=plt.cm.hot)
````

````python
#绘制完整的等高线图
plt.figure(figsize=(10,6))
plt.contourf(X,Y,f(X,Y),8,alpha=0.75,cmap=plt.cm.hot)  #以热力图形式填充每个区域
c=plt.contour(X,Y,f(X,Y),8,colors='black',linewidth=8) #绘制没有标注的等高线轮廓

#绘制等高线数据
plt.clabel(c,inline=True,fontsize=10)        #标注出各等高线的高度
plt.xticks(())
plt.yticks(())    #隐藏x,y坐标轴的数值
plt.show()
````

## 正则表达式 （Regular Expression）

````python
import re
s='123456789'
re_='^2'
m1=re.match(re_,s)
print(m1)
````

````python
m2=re.match('123',s)
m2.span()
````

````python
re1='^1\d{3}'
m3=re.match(re1,s)
print(m3)
````

````python
 #re.match(p,s,[,f])表示按照规则从开始处进行匹配，成
 #参数f表示匹配方式，如re.I表示不区分大小写
 # 功则返回匹配对象，不成功则返回None,返回的对象有.span()
 # 与.group()两个方法，.span()返回匹配的位置，.group()返回匹配的组
 m1 = re.match('123','123456789')
 m1.span()
````

````python
m1.group()
````

````python
m2=re.match('78','123456789')
print(m2)                       #match为从开始处开始匹配，故没匹配到
````

````python
#与match不同，search将会按照规则p搜索整个字符串S,只要从任意位置匹配上就算成
# 功，返回的对象也有.span()与.group(),返回第一次匹配成功的match对象
m3=re.search('456','123456789')
m3.span()
````

````python
m3.group()
````

````python
#re.compile(p,[,f]),返回一个模式对象，这个
# 对象包含了re模块中的绝大多数方法，优点在于
# 可以重复使用同一个正则表达式字符串
p=re.compile('ab')
m5=p.match("absolute")
m6=p.search('enable')
m5.span()
````

````python
m5.group()
````

````python
# re.findall()返回所有匹配成功的子字符串所组成的列表(注意是列表形式)
m7=re.findall('5','152887895438588555')
m7
````

````python
#re.finditer(),这个函数与findall()函数作用类似,但它返回的是一个迭代器，可以用__next__()方法查看
m8=re.finditer('5','12345675657895')
print(m8)
````

````python
#汉诺塔  A是最初位置，C是最终位置
def hanoi(n,A,B,C):
    if n==1:
        print("移动第{}个盘子从{}到{}".format(n,A,C))
    else:
        hanoi(n-1,A,C,B)  #n-1个盘放B,再最大盘移到C
        print("移动第{}个盘子从{}到{}".format(n,A,C))
        hanoi(n-1,B,A,C)

n = int(input("请输入汉诺塔盘子数："))
hanoi(n,"左柱","中柱","右柱")
````

# 闭包

````python
#所谓闭包，就是满足这三种条件的一种函数（1）这个函数中内嵌函数，（2）内嵌函数使用了外层变量（3）这个函数的返回值是内嵌函数
#闭包也是高阶函数的一种表达形式

def func1(x):
    def func2(y):
        return x+y
    return func2                         #注意这里的func2不用加括号

i = func1(3)     #此处返回的是func2,此时i=func2(y)=3+y
n = i(4)
print(n)
````

## 装饰器

````python
#装饰器可以为其他函数锦上添花，使用它可以实现代码的重用
#设计一个程序比较计算1到10求和的时间和求10的阶乘的时间
import time
def get_sum(n):
    startTime = time.clock()
    sum = 0
    for i in range(1,n+1):
        sum+=i
    print("1到{}的正整数和为{}".format(n,sum))
    endTime = time.clock()
    print("程序运行的时间为{}秒".format(endTime-startTime))

def get_fac(n):
    startTime = time.clock()
    fac = 1
    for i in range(1,n+1):
        fac *= i
    print("{}的阶乘为{}".format(n,fac))
    endTime = time.clock()
    print("程序运行的时间为{}秒".format(endTime-startTime))

get_sum(10)
get_fac(10)
````

````python
#如果有多个函数需要这样一个计算求运行时间的功能，每次都写同样的代码显得很繁琐，此时我们久可以加个装饰器了
#装饰器函数
def deco(func):
    def get_time(*args):
        startTime = time.clock()
        func(*args)
        endTime = time.clock()
        print('程序运行的时间是{}秒'.format(endTime-startTime))
    return get_time

#我们只需要在被装饰的函数前面加上@+装饰器函数名即可
@deco
def get_sum(n):
    fac = 1
    for i in range(1,n+1):
        fac *= i
    print("{}的阶乘是{}".format(n,fac))

@deco
def get_sum(n):
    sum = 0
    for i in range(1,1+n):
        sum +=i
    print("1到{}的和为{}".format(n,sum))

get_sum(10)
get_fac(10)
````

# 继承：祖传秘方

````python
#目的是重用代码
#构造人类
#原来的类叫父类，后面继承的类为子类
#定义一个类，如果要继承另一个类，格式如下 class 子类名(父类名):...
class Human():                    #父类
    def __init__(self,name,age):         #构造初始化程序
        self.name = name
        self.age = age
    
    def talk(self):                            #定义了一个talk方法
        print("我是{},我的年龄是{}。".format(self.name,self.age))
    
class Male(Human):
    def __init__(self, name, age):
        super().__init__(name, age)    #子类中不会自动调用父类中的__init__()方法，必须专门调用，使用方法是super().__init__()
        self.sex = "男"                 #添加新的属性
    
    def working(self):
        print("我是{}人，我喜欢工作。".format(self.sex))

class Female(Human):
    def __init__(self, name, age):
        super().__init__(name, age)
        self.sex = "女"

    def shopping(self):
        print("我是个{}人，我喜欢购物。".format(self.sex))

#实例化一个女人
he = Male("Jack",20)
she = Female("Rose",18)
he.talk()
he.working()
she.talk()
she.shopping()
````

## 多继承与多重继承

````python
# 格式 class 子类名(父类1，父类2，...)
#比如我们构造一个人类，在构造一个机器类，用这两个类作为父类，我们再构造一个机器人类，来继承人类和机器类的共同点
#最后实例化一个机器人，看看它能否实现人类和机器的共同功能

class Human():
    def __init__(self,name) :
        self.name = name
    def talk(self):
        print("我的名字叫{}。".format(self.name))

class Machine():
    def __init__(self,function):
        self.function = function
    def work(self):
        print("我在{}".format(self.function))

class Robot(Human,Machine):
    def __init__(self, name,function):
        Human.__init__(self,name)                     #用父类名.__init__()方法来借用父类的多个够着函数
        Machine.__init__(self,function)         #此处因为父类有多个因此不能用upper().__init__()来借用父类的构造函数
    def say(self):
        print("我是一个机器人")
        self.talk()
        self.work()

sweeper = Robot("扫地僧","扫地")
sweeper.say()
````

````python
sweeper.work()    #子类继承了父类的方法，因此父类的方法子类都有
````

````python
#还有一种继承方式，一个子类继承了父类，这个子类又被当作父类继承创建了一个新类，类似子孙三代甚至多带的形式
class Grandfather():
    def __init__(self):
        self.firstname = "王"
    def say1(self):
        print("我姓{}".format(self.firstname))

class Father(Grandfather):
    def __init__(self):
        super().__init__()
        self.money = "一百万"
    def say2(self):
        print("我有{}。".format(self.money))

class Son(Father):
    def __init__(self):
        super().__init__()
    def say(self):
        self.say1()
        self.say2()

#实例化一个Son对象
w = Son()
w.say1()
w.say()
````

# 错误与异常处理

````python
def divide():
    print("计算a除以b的值")
    a = input("请输入a的值：")
    b = input("输入b的值：")

    try:
        c = float(a)/float(b)
    except ZeroDivisionError:
        print("b的值不能为0，请重新输入。")
        divide()
    else:
        print("a除以b的值是{}".format(c))
divide()
````

````python
#在上述代码中，出现异常，则执行异常处理程序，未出现异常，则执行else后面的程序
#实际上以上程序可以不用写else后面的内容
def divide():
    print("计算a除以b的值")
    a = input("请输入a的值：")
    b = input("输入b的值：")

    try:
        c = float(a)/float(b)
        print("a除以b的值是{}".format(c))
    except ZeroDivisionError:
        print("b的值不能为0，请重新输入。")
        divide()     
divide()
````

````python
#其实上述程序还可能出现其他异常，except后面的程序我们可以不用写
#
def divide():
    print("计算a除以b的值")
    a = input("请输入a的值：")
    b = input("输入b的值：")

    try:
        c = float(a)/float(b)
        print("a除以b的值是{}".format(c))
    except:
        print("a和b必须是数字，且b的值不能为0，请重新输入。")
        divide()     
divide()
````

Button

button_one = tk.Button(window,options)

除bg,fg,font外其他常用属性有text，command(按钮所绑定的函数和方法)

````python
#按下ok按钮触发一个事件处理，让之前输出的“Hello world”标签改变文字与背景颜色
import tkinter as tk
window_one = tk.Tk()   #创建主窗口
window_one.title("第一个GUI界面")
window_one.geometry("400x300")  #设置主窗口大小
label_one = tk.Label(window_one,text="Hello world",bg='red')  #创建标签并设置内容和背景颜色
label_two = tk.Label(window_one,text="你好",bg='green',fg='white',font=('楷体',18),width=20)
label_three = tk.Label(window_one,text="世界",bg='blue',fg='black',font=('黑体',20),height=5)

def click_ok():                                      #事件处理函数
    label_one['text'] = "你单击了ok按钮"
    label_one['bg'] = 'yellow'

button_one = tk.Button(window_one,text='OK',command=click_ok) #事件绑定

label_one.pack()   #布局标签
label_two.pack()
label_three.pack()
button_one.pack()

window_one.mainloop()  #执行主事件循环
````

````python
#事实上，tkinter还给各种控提供了一个bind函数专门用来绑定各种事件处理函数，比如我们可以给按钮分别绑定
# 单击事件和双击事件所对应的函数，在定义绑定函数的时候第一个参数默认写为event用于接收事件
#格式：控件.bind(事件，绑定函数)
import tkinter as tk
window_one = tk.Tk()   #创建主窗口
window_one.title("第一个GUI界面")
window_one.geometry("400x300")  #设置主窗口大小
label_one = tk.Label(window_one,text="Hello world",bg='red')  #创建标签并设置内容和背景颜色
label_two = tk.Label(window_one,text="你好",bg='green',fg='white',font=('楷体',18),width=20)
label_three = tk.Label(window_one,text="世界",bg='blue',fg='black',font=('黑体',20),height=5)

def click_ok_once(event):                                      #事件处理函数
    label_one['text'] = "你单击了ok按钮"
    label_one['bg'] = 'yellow'

def click_ok_twice(event):
    label_one['text'] = "你双击了ok按钮"
    label_one['bg'] = 'white'

button_one = tk.Button(window_one,text='OK')  #未在此处绑定事件
button_one.bind('<Button-1>',click_ok_once)
button_one.bind('<Double-1>',click_ok_twice)     

label_one.pack()   #布局标签
label_two.pack()
label_three.pack()
button_one.pack()

window_one.mainloop()  #执行主事件循环
````

##  Entry 输入框控件

创建方法：entry_one = tk.Entry(window,options),除一些常见属性外，还有一些特有的属性

justify:文本对齐方式，默认为left,其他值有center,right

show:输入文本的显示方式，默认是明文，也可以设置为密文，比如用星号"*"

````python
import tkinter as tk
window_one = tk.Tk()
window_one.title("输入用户名和密码")
window_one.geometry('400x300')
label_one = tk.Label(window_one,text='请输入用户名：')
label_two = tk.Label(window_one,text='请输入密码：')
entry_one = tk.Entry(window_one)
entry_two = tk.Entry(window_one,show='*')  #输入内容显示为*

label_one.pack()
entry_one.pack()
label_two.pack()
entry_two.pack()
window_one.mainloop()
````

````python
#Entry控件有一个get()方法，让程序获取用户输入的内容并处理，我们可以在上面的界面上
# 添加一个显示按钮，为这个显示按钮绑定一个显示函数，单击这个按钮以后就可以在显示出用户输入的结果
import tkinter as tk
window_one = tk.Tk()
window_one.title("输入用户名和密码")
window_one.geometry('400x300')
label_one = tk.Label(window_one,text='请输入用户名：')
label_two = tk.Label(window_one,text='请输入密码：')
label_three= tk.Label(window_one,text="待显示区域")
entry_one = tk.Entry(window_one)
entry_two = tk.Entry(window_one,show='*')  #输入内容显示为*

def show_values():
    label_three['text'] = '用户名为{},密码是{}'.format(entry_one.get(),entry_two.get())

button_one = tk.Button(window_one,text='显示',command=show_values)

label_one.pack()
entry_one.pack()
label_two.pack()
entry_two.pack()
button_one.pack()
label_three.pack()
window_one.mainloop()
````

## Scroll滚动条控件，当显示的内容比较长的时候可以增加滚动条控件

格式：scroll_one = tk.Scrollbar(window)

设置滚动条的位置和方向：scroll_one.pack(side,fill) ，参数side表示滚动条的位置，比如tk.RIGHT表示右侧，参数fill,表示滚动条的方向，比如tk.Y表示方向为Y方向

最后需要通过config()方法将滚动条和想把它加入的控件进行绑定

绑定上述文字框内容在右侧添加一个滚动条scroll_one.config(command=text_one.yview)

在创建一个Text控件的时候也需要一个参数来与滚动条控件绑定，text_one.config(yscrollcommand=scrollbar_one.set)

````python
import tkinter as tk
window_one = tk.Tk()
window_one.title("多行文本显示")
window_one.geometry('400x200')
scrollbar_one = tk.Scrollbar(window_one)              #创建滚动条控件
scrollbar_one.pack(side=tk.RIGHT,fill=tk.Y)            #设置滚动条位置与方向

text_one = tk.Text(window_one,yscrollcommand=scrollbar_one.set)      #创建文本框，并与滚动条绑定
text_one.pack()

scrollbar_one.config(command=text_one.yview)      #函数绑定
poem= r'''《短歌行》
曹操
对酒当歌，人生几何！
譬如朝露，去日苦多。
慨当以慷，忧思难忘。
何以解忧？唯有杜康。
青青子衿，悠悠我心。
但为君故，沉吟至今。
呦呦鹿鸣，食野之苹。
我有嘉宾，鼓瑟吹笙。
明明如月，何时可掇？
忧从中来，不可断绝。
越陌度阡，枉用相存。
契阔谈䜩，心念旧恩。
月明星稀，乌鹊南飞。
绕树三匝，何枝可依？
山不厌高，海不厌深。
周公吐哺，天下归心。
'''
text_one.insert(tk.INSERT,poem)
window_one.mainloop()
````

## Radiobutton 单选按钮

````python
import tkinter as tk
window_one = tk.Tk()
window_one.title('单选按钮')
window_one.geometry('400x200')
label_one = tk.Label(window_one,text='一加一等于几')
label_two = tk.Label(window_one,text='答题结果') 
v= tk.IntVar()         #v是一个整数类型变量对象，可以用.get()方法获取
def show():
    if v.get()==1:
        label_two['text']='你数学是体育老师教的吗？错'
    elif v.get()==2:
        label_two['text']='恭喜你答对了'
    else:
        label_two['text']='你喝醉了吗？错'
radiobutton_one = tk.Radiobutton(window_one,text='一',variable=v,value=1,command=show)
radiobutton_two = tk.Radiobutton(window_one,text='二',variable=v,value=2,command=show)
radiobutton_three = tk.Radiobutton(window_one,text='三',variable=v,value=3,command=show)
label_one.pack()
radiobutton_one.pack()
radiobutton_two.pack()
radiobutton_three.pack()
label_two.pack()
window_one.mainloop()
````

## Checkbutton 多选按钮，这个控件可以创建一个多选按钮，往往是一组同时使用

````python
import tkinter as tk
w_one = tk.Tk()
w_one.title("多选按钮")
w_one.geometry('400x200')
l_one = tk.Label(w_one,text='下列那几首歌是周杰伦唱的？')
l_two = tk.Label(w_one,text='答题结果')
v1 = tk.IntVar()
v2 = tk.IntVar()
v3 = tk.IntVar()
v4 = tk.IntVar()
def show():
    if v1.get()==v3.get()==1 and v2.get()==v4.get()==0:
        l_two['text']='恭喜你答对了'
    else:
        l_two['text']='看来你不是周杰伦的歌迷....'

cb_one=tk.Checkbutton(w_one,text='上海一九四三',variable=v1,onvalue=1,offvalue=0)      #onvalue可以设置选项被选中时候的值，offvalue可以设置选项没被选中时的值
cb_two=tk.Checkbutton(w_one,text='铁岭一九四三',variable=v2,onvalue=1,offvalue=0)
cb_three=tk.Checkbutton(w_one,text='爸我回来了',variable=v3,onvalue=1,offvalue=0)
cb_four=tk.Checkbutton(w_one,text='二大爷我回来了',variable=v4,onvalue=1,offvalue=0)

b_one = tk.Button(w_one,text='提交答案',command=show)
l_one.pack()
cb_one.pack()
cb_two.pack()
cb_three.pack()
cb_four.pack()
b_one.pack()
l_two.pack()
w_one.mainloop()
````
